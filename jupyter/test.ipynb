```python
url = 'http://quote.eastmoney.com/stocklist.html'
```

```python
import requests
```

```python
def getHTMLText(url):
    try:
        r = requests.get(url)
        r.raise_for_status()
        r.encoding = r.apparent_encoding
        return r.text
    except:
        return ""
```

```python
html = getHTMLText(url)
```

```python
print(html)
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
a = soup.find_all('a')
print(a)
```

```python
import re
lst = []
for i in a:
    try:
        href = i.attrs['href']
        stock = re.findall(r"[s][hz]\d{6}",href)[0]
        lst.append(stock)
#         print(stock)
    except:
        continue
print(lst)
#     stock = re.findall(r"[s][hz]\d{6}",href)[0]
#     print(href)
```

```python
stockURL = 'https://gupiao.baidu.com/stock/'
for stock in lst:
    url = stockURL + stock + '.html'
    print(url)
```
