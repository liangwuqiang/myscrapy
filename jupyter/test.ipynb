```python
url = 'http://www.maiziedu.com/wiki/crawler/introduce/'
```

```python
import requests
```

```python
res = requests.get(url)
```

```python
res.encoding = res.apparent_encoding
```

```python
html = res.text
```

```python
print(html)
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
hrefs = soup.select('.wikiPageNavMenu > .navBar > li > a')
print(len(hrefs))
for href in hrefs:
    print(href['href'])
```
